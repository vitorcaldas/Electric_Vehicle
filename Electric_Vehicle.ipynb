```python
# Contar o número de veículos por ano.
year_counts = df['Year'].value_counts().sort_index().reset_index()
year_counts.columns = ['Year', 'Vehicle_Count']

plt.figure(figsize=(10, 6))
sns.barplot(data=year_counts, x='Year', y='Vehicle_Count', palette='mako')
plt.title('Número de veículos por ano', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Agrupe por marca e calcule a média de CO2 economizado.
co2_by_make = df.groupby("Make")["CO2_Saved_tons"].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
co2_by_make.plot(kind="bar", color="seagreen")
plt.title("Economia média de CO₂ (toneladas) por produção")
plt.ylabel("CO₂ Saved (tons)")
plt.xlabel("Make")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Modelagem preditiva

```python
y = df['Monthly_Charging_Cost_USD']
X = df.drop(['Monthly_Charging_Cost_USD', 'Region', 'Usage_Type', 'CO2_Saved_tons', 'Insurance_Cost_USD', 'Temperature_C', 'Vehicle_ID'], axis=1)
```

```python
X['Charging_Power_kW'] = pd.to_datetime(X['Charging_Power_kW'])
X['Charging_Time_hr'] = X['Charging_Power_kW'].dt.year
X['Charge_Cycles'] = X['Charging_Power_kW'].dt.month
X['Electricity_Cost_USD_per_kWh'] = X['Charging_Power_kW'].dt.day
X = X.drop('Charging_Power_kW', axis=1)
```

```python
categorical_cols = X.select_dtypes(include=["object"]).columns.tolist()
```

```python
X_encoded = pd.get_dummies(X, columns=categorical_cols)
print(X_encoded.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")
print(f"R² Score: {r2:.2f}")
```

Prever o valor de revenda dos veículos elétricos

```python
y = df['Resale_Value_USD']
X = df.drop(['Resale_Value_USD', 'Vehicle_ID'], axis=1)

print("Target variable 'y' head:")
print(y.head())
print("\nFeatures 'X' head:")
print(X.head())
```

```python
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

print("Categorical columns identified:")
print(categorical_cols)
print("\nFeatures 'X_encoded' head after one-hot encoding:")
print(X_encoded.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

print("Shapes of the datasets after splitting:")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

models = {
    "Linear Regression": LinearRegression(),
    "Random Forest Regressor": RandomForestRegressor(random_state=42),
    "Gradient Boosting Regressor": GradientBoostingRegressor(random_state=42)
}

results = {}

for name, model in models.items():
    print(f"\nTraining {name}...")

    model.fit(X_train, y_train)


    y_pred = model.predict(X_test)


    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)


    results[name] = {"MSE": mse, "RMSE": rmse, "R2 Score": r2}


    print(f"{name} Performance:")
    print(f"  Mean Squared Error (MSE): {mse:.2f}")
    print(f"  Root Mean Squared Error (RMSE): {rmse:.2f}")
    print(f"  R\u00b2 Score: {r2:.2f}")
```

```python
print("\n--- Linear Regression Feature Importance ---")
linear_reg_model = models["Linear Regression"]
coefficients = pd.DataFrame({
    'Feature': X_encoded.columns,
    'Coefficient': linear_reg_model.coef_
})
coefficients['Abs_Coefficient'] = np.abs(coefficients['Coefficient'])
coefficients = coefficients.sort_values(by='Abs_Coefficient', ascending=False)
print("Top 10 features for Linear Regression (by absolute coefficient magnitude):\n", coefficients.head(10))
```

```python
print(
    "\n--- Random Forest Regressor Feature Importance ---"
)
rf_model = models["Random Forest Regressor"]
feature_importances_rf = pd.DataFrame({
    'Feature': X_encoded.columns,
    'Importance': rf_model.feature_importances_
})
feature_importances_rf = feature_importances_rf.sort_values(
    by='Importance', ascending=False
)
print(
    "Top 10 features for Random Forest Regressor: \n",
    feature_importances_rf.head(10),
)
```

```python
print(
    "\n--- Gradient Boosting Regressor Feature Importance ---"
)
gbr_model = models["Gradient Boosting Regressor"]
feature_importances_gbr = pd.DataFrame({
    'Feature': X_encoded.columns,
    'Importance': gbr_model.feature_importances_
})
feature_importances_gbr = feature_importances_gbr.sort_values(
    by='Importance', ascending=False
)
print(
    "Top 10 features for Gradient Boosting Regressor: \n",
    feature_importances_gbr.head(10),
)
```

## Recomendações de Negócio







   * **Desempenho do Modelo:** Todos os três modelos de regressão (Regressão Linear, Regressor de Floresta Aleatória e Regressor de Gradiente Impulsionado) demonstraram forte capacidade preditiva, com valores de R² em torno de 0,90, indicando que explicam aproximadamente 90% da variância no valor de revenda de veículos elétricos.
    *   Linear Regression: Achieved an R² of 0.90, MSE of 3,108,489.06, and RMSE of 1,763.09.
    *   Random Forest Regressor: Achieved an R² of 0.89, MSE of 3,301,572.57, and RMSE of 1,817.02.
    *   Gradient Boosting Regressor: Achieved an R² of 0.90, MSE of 3,160,360.56, and RMSE of 1,777.74.
    
   A regressão linear e o regressor Gradient Boosting superaram ligeiramente o regressor Random Forest em termos de pontuação R² e RMSE.

* **Principais fatores que influenciam o valor de revenda:**

* **Capacidade e idade da bateria (anos):** Para os modelos baseados em árvores (Random Forest e Gradient Boosting), 'Capacidade_da_bateria_kWh' foi de longe a característica mais importante (pontuações de importância de 0,85 e 0,91, respectivamente), seguida por 'Ano' (pontuações de importância de 0,057 e 0,066, respectivamente). Isso sugere que o tamanho da bateria e a idade do veículo são os principais determinantes do valor de revenda.

* **Impacto Ambiental e Custos Operacionais:** Na regressão linear, 'CO2_Saved_tons' apresentou o maior coeficiente negativo absoluto (-$25.229,93), O que implica que uma maior quantidade de CO2 economizado (potencialmente indicando um modelo de veiculo mais antigo ou menos eficiente ao longo de sua vida útil) pode estar associada a um menor valor de revenda. 'Electricity_Cost_USD_per_KWh' apresentou o coeficiente positivo ($$900,92), sugerindo custos de eletricidade mais baixos por kWh estao associados a valores de revenda mais altos.

* **Modelos Específicos:** Certos modelos de carros, como o 'Model_Model S', mostraram uma influência positiva no valor de revenda, enquanto o 'Model_Model 3' mostrou uma influência negativa no modelo de regressão linear.

### Insights ou Próximos Passos

* **Priorizar a Saúde e a Capacidade da Bateria:** Fabricantes e concessionárias devem enfatizar a capacidade da bateria e, potencialmente, oferecer garantias estendidas ou certificados de saúde da bateria para otimizar o valor de revenda de veículos elétricos, já que este é o fator dominante.

* **Destacar o Ano do Veículo e os Avanços Tecnológicos:** Modelos mais novos ('Ano') retêm melhor o valor, o que é esperado. O marketing deve focar nos avanços tecnológicos recentes e na melhoria da eficiência dos modelos mais novos para justificar preços de revenda mais altos.
